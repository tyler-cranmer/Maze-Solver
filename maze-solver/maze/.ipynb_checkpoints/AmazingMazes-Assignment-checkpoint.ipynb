# Assignment: Solving Image Maze 

Given a maze as an image with a start and end point, we would like to write code to solve the maze. 

An image is a 2D matrix of pixels of a particular size that depends on its resolution. Each 
pixel has a color which is given by its Red, Green and Blue (RGB) values.
 
Given an image, we will view it as a graph where each pixel of the image is a vertex 
and edges connect a pixel to its neighbor. The weight of an edge should be very small if
the pixel colors are similar (i.e, the differences between r, g and b values are close to zero)
and correspondingly large as the pixel colors diverge.
  
Next, given a source pixel $(i_0, j_0)$ and destination pixel, 
$(i_1, j_1)$, we wish find the shortest weight path from source to destination.

You should use the Dijkstra's algorithm modified in two ways:
   - It can exit as soon as the destination is reached.
   - A 1000 x 1000 pixel image gives rise to a graph with million vertices. Storing such a graph as an adjacency list is going to be very memory intensive. Instead, your goal will be to generate the vertices and edges *on-the-fly*.
   
We will use opencv library, a popular computer vision library to load, and manipulate images of mazes.

The pixel color is expressed in RGB format. R is the red value from 0 -> 255, G is the green value 0 -> 255 and B is the blue value from 0 -> 255 

We will now define a edge weight function for an edge in the image

```python
import math 
def fixPixelValues(px):
    # convert the RGB values into floating point to avoid an overflow that will give me wrong answers
    return [ float(px[0]), float(px[1]), float(px[2]) ]

## Given (x,y) coordinates of two neighboring pixels, calculate the edge weight.
# We take the squared euclidean distance between the pixel values and add 0.1
def getEdgeWeight(img, u, v):
    # get edge weight for edge between u, v
    # First make sure that the edge is legit
    i0,j0 = u[0], u[1]
    i1,j1 = v[0], v[1]
    height, width, _ = img.shape
    assert i0 >= 0 and j0 >= 0 and i0 < width and j0 < height # pixel position valid?
    assert i1 >= 0 and j1 >= 0 and i1 < width and j1 < height # pixel position valid?
    assert -1 <= i0 - i1 <= 1 # edge between node and neighbor?
    assert -1 <= j0 - j1 <= 1
    px1 = fixPixelValues(img[i0,j0]) #(img(u[0],u[1]))
    px2 = fixPixelValues(img[i1,j1]) #(img[v[0],v[1]])
    return 0.1 + (px1[0] - px2[0])**2 + (px1[1] - px2[1])**2 + (px1[2]- px2[2])**2

# This is a useful function that given a list of (x,y) values, 
# draw a series of red lines between each coordinate and next to 
# show the path in the image
def drawPath(img, path, pThick=2):
    v = path[0]
    x0, y0 = v[0], v[1]
    for v in path:
        x, y = v[0], v[1]
        cv2.line(img,(x,y), (x0,y0), (255,0,0),pThick)
        x0, y0 = x,y
        
```

## Step 1: Compute Single Source Shortest Path For an Image

Given an image, compute the  shortest path between source and destination pixels by modifying Dijkstra's algorithm. __Your challenge  is to implement it without needing to create the entire the adjacency list for the graph
in the first place. However, for simplicity you can try a first cut implementation of a generic Dijkstra algorithm over
graphs represented as adjacency matrix or list.__ 

```python
# Vertex and Priority Queue classes.

class Vertex: # This is the outline for a vertex data structure
    def __init__ (self,  x_cord, y_cord):
        self.x = x_cord # The x coordinate
        self.y = y_cord  # The y coordinate
        self.d = float('inf') # the shortest path estimate
#         self.d = random.randrange(0, 5000) #<----------------------------------------------------------- for testing
        self.processed = False # Has this vertex's final shortest path distance been computed
        self.in_queue = False # is the vertex currently in the queue?
        self.idx_in_priority_queue = -1 # The index of this vertex in the queue
        self.pi_x = None 
        self.pi_y = None # the parent vertex.

        
class PriorityQueue:
    # Constructor: 
    def __init__(self):
        self.q = [] #priority queue

    def insert(self, vertex): #insert
        vertex.idx_in_priority_queue = len(self.q) #set vertex index to the last position in queue
        vertex.in_queue = True
        self.q.append(vertex) #insert vertex to queue
        self.bubble_up(vertex.idx_in_priority_queue)  #bubble up vertex to fix min heap structure

    def bubble_up(self, index): #bubble up to fix heap for insert
        parent = (index-1) // 2  #parent node
        child = index #child node
        if index <= 0:
            return 
        if(self.q[child].d < self.q[parent].d): #if child distance is < parents distance; swap positions
            self.q[child],self.q[parent] = self.q[parent],self.q[child] #swap
            self.q[child].idx_in_priority_queue = child #update priority number for child
            self.q[parent].idx_in_priority_queue = parent #update priority number for parent
            self.bubble_up(parent) #bubble up to parent
        return
    
    def extract_min(self):         #extracts min vertex from heap, shortens the array after each iteraton
        qSize = len(self.q) - 1 # set queue size
        first_vertex = self.q[0] #grab first vertex from queue
        first_vertex.processed = True # vertex has been processed
        first_vertex.in_queue = False
        self.q[0] = self.q[qSize] #swap the last vertex in queue to top of queue
        self.q[0].idx_in_priority_queue = 0 #set the new top vertex index to 0
        self.q.pop(qSize) # remove last vertex
        self.bubble_down(0) #bubble down top of queue
    
        return first_vertex

    def bubble_down(self, index): # takes in array and element to be analyzed and bubble down to create min heap. heapify
        smallest = index     
        left = 2 * index + 1 #left child
        right = left + 1 #right child
        length = len(self.q) 
        if(left < length and self.q[left].d < self.q[smallest].d): 
            smallest = left                                        
        else:
            smallest = index
        if(right < length and self.q[right].d < self.q[smallest].d):
            smallest = right
        if smallest != index:
            self.q[index],self.q[smallest] = self.q[smallest],self.q[index] #swap
            self.q[smallest].idx_in_priority_queue = smallest #update priority number
            self.q[index].idx_in_priority_queue = index #update priority number
            self.bubble_down(smallest)
    
    def show(self): #prints whats in the queue
        for vertex in self.q:
            print("Vertex:(%s,%s)" % (vertex.x, vertex.y))
            print("distance value index: %f" %  vertex.d)
            print("priority number = %d" % vertex.idx_in_priority_queue)
            print("parent = (%s, %s)" % (vertex.pi_x , vertex.pi_y))
            print("processed : %s \n" % vertex.processed)
    
    def show_one(self): #prints first vertex in queue
        print("Vertex:(%s,%s)" % (self.q[0].x, self.q[0].y))
        print("distance value index: %f" %  self.q[0].d)
        print("priority number = %d" % self.q[0].idx_in_priority_queue)
        print("parent = (%s, %s)" % (self.q[0].pi_x , self.q[0].pi_y))
        print("processed : %s \n" % self.q[0].processed)
    
    def is_empty(self): #return true or false if queue is empty. 
        return len(self.q) == 0
```

```python
def get_neighbors(mat,row,col): #retrieves the neighboring vertices from a current node
    shape=mat.shape
    neighbors=[]
    #check to make sure correct neighboring vertices
    if row > 0 and not mat[row-1][col].processed: 
         neighbors.append(mat[row-1][col]) #bottom neighboring vertex
    if row < shape[0] - 1 and not mat[row+1][col].processed: 
            neighbors.append(mat[row+1][col]) #top  neighboring vertex
    if col > 0 and not mat[row][col-1].processed: 
        neighbors.append(mat[row][col-1])#left neighboring vertex
    if col < shape[1] - 1 and not mat[row][col+1].processed: 
            neighbors.append(mat[row][col+1]) #right neighboring vertex
    return neighbors
```

```python
def computeShortestPath(img, source, dest): 
    pq = PriorityQueue() #initialize priority queue
    path = [] # list for the minimum path
    
    start_x = source[0] #starting coordinates
    start_y = source[1] #starting coordinates
    
    end_x = dest[0] #ending coordinates
    end_y = dest[1] #ending coordinates
    
    
    imagerows = img.shape[0] 
    imagecols = img.shape[1] #setting rows and cols of image
    matrix = np.full((imagerows, imagecols), None)  #access by matrix[row][col]
    
    for rows in range(imagerows):
        for cols in range(imagecols):
            matrix[rows][cols] = Vertex(cols,rows)   #set image matrix to vertices
    
    matrix[start_y][start_x].d=0 #set starting location to distance 0
    pq.insert(matrix[start_y][start_x]) #insert source into queue

    
    while pq.is_empty() is not True: #while Que is not empty
        u = pq.extract_min()   #extract min vertex in queue
        neighbors = get_neighbors(matrix, u.y, u.x) # create a list of all neighboring vertex to u
        for v in neighbors: #iterate through all neighbors
                dist = getEdgeWeight(img, (u.y, u.x), (v.y, v.x)) #calculate thier distances
                if v.in_queue is not True: #if neighbor is not in the queue: add to queue
                    pq.insert(v)
                if u.d + dist < v.d: #if current node + distance is less than neighbor's distance
                    v.d = u.d + dist #set neighbor distance to new larger distance
                    v.pi_x= u.x # set current node as neighbors new parent vertex
                    v.pi_y = u.y # set current node as neighbors new parent vertex
                    pq.bubble_up(v.idx_in_priority_queue) #bubble up neighbor vertex
                
    current_node = matrix[end_y][end_x]  #set destination vertex as current node
    path.append((end_x,end_y))           #append destination coordinate to path
    while(current_node.y != start_y or current_node.x != start_x): #while the curren_node's coordinates are not the starting coordinates
        path.append((current_node.x,current_node.y))              #append nodes x and y coordinates
        current_node=matrix[current_node.pi_y][current_node.pi_x] #set current_node to the parent of last node.
        
    path.append((start_x,start_y)) #finally append the source coordinates
    return path 
```
